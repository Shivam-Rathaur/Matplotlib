Plotting Without Line

##### Format Strings `fmt`

* We can also use the `shortcut string notation parameter` to specify the marker, line style, and color in plotting.
* This `parameter` is also called `fmt`, and is written with this `syntax`:

`marker|line|color`

```python
'''
    In matplotlib, the fmt parameter is a shorthand notation used to specify 
    the marker, line style, and color in plotting functions like plt.plot(). 
'''

# plt.plot(x, y, 'marker|line|color')

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

## 'o--r' -> fmt i.e., format string
plt.plot(x, y, 'o--r')  # circle markers(o), dashed line(--), red color(r)
plt.show()
```

```python
## Marker Color
# You can use the keyword argument `markeredgecolor`` or the shorter `mec`` to set the color of the edge of the markers:
# You can use the keyword argument `markerfacecolor`` or the shorter `mfc`` to set the color inside the edge of the markers:
```

```python
## Set the EDGE color to red:
x_points = np.array([1,2,3,4,5,6])
y_points = np.array([10,20,30,40,50,60])

plt.plot(x_points, y_points, '*:c', ms=20, mec='r')
plt.show()
```

```python
## Set the FACE color to red:

x_points = np.array([1,2,3,4,5,6])
y_points = np.array([10,20,30,40,50,60])

plt.plot(x_points, y_points, '*:c', ms=20, mfc='r')
plt.show()
```

```python
## Set the color of both the edge and the face to red:

x_points = np.array([1,2,3,4,5,6])
y_points = np.array([10,20,30,40,50,60])

plt.plot(x_points, y_points, '*:c', ms=20, mec='r', mfc='r')
plt.show()
```

Set Font Properties for Title and Labels

`fontdict` parameter in `xlabel(), ylabel(), and title()` to set font properties for the title and labels.
- The fontdict parameter in matplotlib stands for Font Dictionary.

```python
## Example
# Set font properties for the title and labels:

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif', 'color':'red', 'fontsize':12}
font2 = {'family':'serif', 'color':'green', 'fontsize':10}

plt.title('Sports Watch Data', fontdict=font2)
plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel('Calorie Burnage', fontdict = font1)

plt.plot(x,y)
plt.show()
```

```python
## position of title: `loc` [stands for location] parameter in title() to position the title
# Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif', 'color':'red', 'fontsize':12}
font2 = {'family':'serif', 'color':'green', 'fontsize':10}

plt.title('Sports Watch Data', fontdict=font2, loc='right')
plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel('Calorie Burnage', fontdict = font1)

plt.plot(x,y)
plt.show()
```

#### Subplot: Display Multiple Plots
- With the `subplot()` function you can draw multiple plots in one figure:

The subplot() function

- The subplot() function takes three arguments that describes the layout of the figure.
- The layout is organized in rows and columns, which are represented by the first and second argument.
- The third argument represents the index of the current plot.

plt.subplot(1, 2, 1)
- the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2)
- the figure has 1 row, 2 columns, and this plot is the second plot.


```python
## Draw 2 plots side by side

# plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1,2,1)
plt.plot(x,y)

# plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
## Draw 2 plots on top of each other

# plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2,1,1)
plt.plot(x,y)

# plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

```python
## We can draw as many plots you like on one figure, just descibe 
# the number of rows, columns, and the index of the plot.

# Plot 1
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([30, 40, 50, 10, 20])
plt.subplot(2,3,1)
plt.plot(x1,y1)

# Plot 2
x2 = np.array([6, 7, 8, 9, 10])
y2 = np.array([45, 55, 35, 25, 35])
plt.subplot(2,3,2)
plt.plot(x2,y2)

# Plot 3
x3 = np.array([11, 12, 13, 14, 15])
y3 = np.array([5, 15, 25, 35, 45])
plt.subplot(2,3,3)
plt.plot(x3,y3)

# Plot 4
x4 = np.array([16, 17, 18, 19, 20])
y4 = np.array([50, 40, 30, 20, 10])
plt.subplot(2,3,4)
plt.plot(x4,y4)

# Plot 5
x5 = np.array([21, 22, 23, 24, 25])
y5 = np.array([60, 70, 80, 90, 10])
plt.subplot(2,3,5)
plt.plot(x5,y5)


# Plot 6
x6 = np.array([26, 27, 28, 29, 30])
y6 = np.array([36, 48, 60, 12, 24])
plt.subplot(2,3,6)
plt.plot(x6,y6)

plt.show()
```

Creating Scatter Plots
* `scatter()` function to draw a scatter plot

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) # age
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) # speed
plt.scatter(x,y)
plt.show()
```

```python
## Compare Plots

# Draw two plots on the same figure:

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) # age 
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) # speed
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12]) # age
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85]) # speed
plt.scatter(x, y)

plt.show()
```

- `color` or `c` argument: used to set the color of the scatter plot.

```python
## Example
# Set your own color of the markers

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) # age 
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) # speed
plt.scatter(x, y, color='hotpink')

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12]) # age
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85]) # speed
plt.scatter(x, y, color='c')

plt.show()
```

```python
## color each dot
## to set a specific color for each dot by using an array of colors as value for the `c` argument:

# Note: You cannot use the `color` argument for this, only the `c` argument.
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x,y,c=colors)
plt.show()
```

ColorMap: A colormap (`cmap` -> parameter) is like a list of colors, where each color has a value that ranges from 0 to 100.
- 'viridis' is one of the built-in colormaps available in Matplotlib.
- 'viridis' is the default colormap used to represent data visually in plots like scatter, imshow, contour, etc.
-  'viridis' ranges from 0, which is a purple color, up to 100, which is a yellow color.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y,c=colors,cmap='viridis')
plt.show()
```

```python
## plt.colorbar(): to include the colormap in the figure/graph

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

### Pie Charts

`pie()` function to draw pie charts

```python
# A simple pie chart

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

- pie chart draws one piece (called a `wedge`) for each value in the array (in this case [35, 25, 25, 15]).
- By default the plotting of the first wedge starts from the `x-axis` and moves `counterclockwise`.

```python
## Note: The size of each wedge is determined by comparing the value with all the other values, by using this formula:
# The value divided by the sum of all values: `x/sum(x)`
```

`labels` parameter: add labels to the pie chart.
- The labels parameter must be an array with one label for each wedge

```python
## Ex: A simple pie chart

x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(x, labels=mylabels)
plt.show()
```

`startangle` parameter: to change the start angle (default angle is 0 from x-axis measured acticlockwise).


```python
x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(x, labels=mylabels, startangle=90)
plt.show()
```

- `explode` parameter: to stand out one of the wedges.
- explode parameter must be an array with one value for each wedge. (by default is None)

```python
## Pull the "Apples" wedge 0.2 from the center of the pie

x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(x, labels=mylabels, explode=myexplode)
plt.show()
```

- `shadow` parameter: to add a shadow to the pie chart.

```python
x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(x, labels=mylabels, explode=myexplode, shadow=True)
plt.show()
```

- `colors` parameter: to set the color of each wedge

```python
x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ['red', 'cyan', 'green', 'hotpink']

plt.pie(x, labels=mylabels, explode=myexplode, shadow=True, colors=mycolors)
plt.show()
```

`legend()` function: to add a list of explanation for each wedge.

```python
x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ['red', 'cyan', 'green', 'hotpink']

plt.pie(x, labels=mylabels, explode=myexplode, shadow=True, colors=mycolors)
plt.legend()
plt.show()
```

Note: To add a `header` to the `legend`, add the `title parameter` to the `legend function`.

```python
x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ['red', 'cyan', 'green', 'hotpink']

plt.pie(x, labels=mylabels, explode=myexplode, shadow=True, colors=mycolors)
plt.legend(title='Fruits')
plt.show()
```
